```python
df["Upholstery_Cloth"] = pd.Series()
df["Upholstery_Color"] = pd.Series()
```

```python
for i in range(df.shape[0]):
  if (type(df["Upholstery"][i]) == str) and (len(df["Upholstery"][i].split(",")) > 1):
    df["Upholstery_Cloth"][i] = df["Upholstery"][i].split(",")[0]
    df["Upholstery_Color"][i] = df["Upholstery"][i].split(",")[1]

  elif (type(df["Upholstery"][i]) == str) and (len(df["Upholstery"][i].split(",")) == 1):
    if df["Upholstery"][i] in df["Upholstery_Cloth"].unique():
      df["Upholstery_Cloth"][i] = df["Upholstery"][i].split(",")[0]
    elif df["Upholstery"][i] in df["Upholstery_Color"].unique():
      df["Upholstery_Color"][i] = df["Upholstery"][i].split(",")[0]
```
